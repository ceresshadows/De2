```python
import numpy as np
import csv
import pandas as pd
import re
```

```python
def npc_from_txt(filepath):
    f = open(filepath, 'r')
    fframe = f.readlines()
    
    #append timestamp
    df_ts = pd.DataFrame(columns=['npc_ts'])
    #append npc
    df_lc = pd.DataFrame(columns=['npc_x','npc_y','theta'])
    
    for row in fframe:
        if row.find('timestamp_sec') != -1:
            ts = re.findall("(?<=[AZaz])?[0-9.+-]+",row) #exclude "timestamp_sec:"
            ts = float(ts[0]) #convert to data format
            l_no = fframe.index(row)
            
            lc_x = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no + 11])
            lc_x = float(lc_x[0])
            lc_y = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no + 12])
            lc_y = float(lc_y[0])
            theta = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no +15])
            theta = float(theta[0])
            
            d_0 = {'npc_ts':[ts]}
            d_df0 = pd.DataFrame(data=d_0)
            df_ts = pd.concat([df_ts, d_df0])

            d_1 = {'npc_x':[lc_x], 'npc_y':[lc_y],'theta':[theta]}
            d_df1 = pd.DataFrame(data=d_1)
            df_lc = pd.concat([df_lc, d_df1])
    df = pd.concat([df_ts, df_lc], axis=1)

    return df
```

```python
def ego_from_txt(filepath):
    f = open(filepath, 'r')
    fframe = f.readlines()
    
    #append timestamp
    df_ts = pd.DataFrame(columns=['ego_ts'])
    #append position
    df_lc = pd.DataFrame(columns=['position_x','position_y','heading'])
    
    for row in fframe:
        if row.find('timestamp_sec') != -1:
            ts = re.findall("(?<=[AZaz])?[0-9.+-]+",row) #exclude "timestamp_sec:"
            ts = float(ts[0]) #convert to data format
            l_no = fframe.index(row)
            
            lc_x = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no + 6])
            lc_x = float(lc_x[0])
            lc_y = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no + 7])
            lc_y = float(lc_y[0])
            heading = re.findall("(?<=[AZaz])?[0-9.+-]+",fframe[l_no +31])
            heading = float(heading[0])
            
            d_0 = {'ego_ts':[ts]}
            d_df0 = pd.DataFrame(data=d_0)
            df_ts = pd.concat([df_ts, d_df0])

            d_1 = {'position_x':[lc_x], 'position_y':[lc_y],'heading':[heading]}
            d_df1 = pd.DataFrame(data=d_1)
            df_lc = pd.concat([df_lc, d_df1])
    df = pd.concat([df_ts, df_lc], axis=1)

    return df
```

```python
def cmd_from_txt(filepath):
    f = open(filepath, 'r')
    fframe = f.readlines()
    
    # Initialize an empty DataFrame
    df = pd.DataFrame(columns=['cmd_ts', 'throttle', 'brake', 'steering'])
    
    for l_no, row in enumerate(fframe):
        if row.strip() == "header {":
            ts = re.findall("(?<=[AZaz])?[0-9.+-]+", fframe[l_no + 1])
            ts = float(ts[0])

            throttle = re.findall("(?<=[AZaz])?[0-9.+-]+", fframe[l_no + 11])
            throttle = float(throttle[0])

            brake = re.findall("(?<=[AZaz])?[0-9.+-]+", fframe[l_no + 12])
            brake = float(brake[0])

            steering = re.findall("(?<=[AZaz])?[0-9.+-]+", fframe[l_no + 14])
            steering = float(steering[0])

            temp_df = pd.concat([pd.DataFrame({'cmd_ts': [ts]}), 
                                 pd.DataFrame({'throttle': [throttle]}), 
                                 pd.DataFrame({'brake': [brake]}), 
                                 pd.DataFrame({'steering': [steering]})], axis=1)
            
            df = pd.concat([df, temp_df], ignore_index=True)

    return df
```

```python
def align_and_average(serial):
    df_npc = npc_from_txt(f'assets/data/{serial}_prediction.txt')
    df_ego = ego_from_txt(f'assets/data/{serial}_pose.txt')
    df_cmd = cmd_from_txt(f'assets/data/{serial}_control.txt')

    ego = df_ego.to_numpy(dtype = np.float64)
    npc = df_npc.to_numpy(dtype = np.float64)
    cmd = df_cmd.to_numpy(dtype = np.float64)
    
    # 初始化结果数组
    aligned_ego = np.zeros_like(npc)
    avg_cmd = np.zeros_like(npc)
    
    # 上一个npc时间戳的索引
    last_npc_idx = 0
    
    for i in range(npc.shape[0]):
        # 对齐ego的时间戳
        ego_idx = np.where(ego[:, 0] < npc[i, 0])[0][-1]
        aligned_ego[i] = ego[ego_idx]
        
        # 平均cmd的数据
        cmd_indices = np.where((cmd[:, 0] >= npc[last_npc_idx, 0]) & (cmd[:, 0] < npc[i, 0]))[0]
        avg_cmd[i] = np.mean(cmd[cmd_indices], axis=0)
        
        last_npc_idx = i

    return aligned_ego, avg_cmd

# 假设 ego, npc, cmd 是你的三个numpy数组，它们的第一列是时间戳   
```

```python
aligned_ego, avg_cmd = align_and_average(serial = 1)
```
